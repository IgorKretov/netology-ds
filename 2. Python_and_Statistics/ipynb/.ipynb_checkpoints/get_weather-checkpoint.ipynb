```python
import pandas as pd
info = pd.read_html('https://yandex.ru/pogoda/moscow/details#19',encoding='utf-8',)[0:3]

today = info[0]
today.columns = ["desc","pic","type","pressure","humidity","wind","real"]
today.drop("pic",axis=1,inplace=True)

tomorrow = info[1]
tomorrow.columns = ["desc","pic","type","pressure","humidity","wind","real"]
tomorrow.drop("pic",axis=1,inplace=True)

after_tomorrow = info[2]
after_tomorrow.columns = ["desc","pic","type","pressure","humidity","wind","real"]
after_tomorrow.drop("pic",axis=1,inplace=True)
```
