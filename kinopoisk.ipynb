```python
from kinopoisk.movie import Movie
from kinopoisk.person import Person
```

```python
print('create table films( \n\
         title varchar(255) -- название (текст) \n\
        ,id serial --соответствует film_id в табличке persons2content \n\
        ,country varchar(1000) --страна \n\
        ,box_office serial --сборы в долларах \n\
        ,release_year timestamp --год выпуска \n\
    );\n')

print('create table persons( \n\
        id serial -- соответствует person_id в табличке persons2content \n\
        ,fio varchar(255)  --фамилия, имя \n\
        );\n')

print('create table persons2content( \n\
        person_id serial -- id персоны \n\
        ,film_id serial -- id контента \n\
        ,person_type varchar(255) -- тип персоны (актёр, режиссёр и т.д.) \n\
     );\n')
```

```python
movie_list = ['Большой куш','Граф Монте-Кристо','Перевозчик','Угнать за 60 секунд','I Am Legend','Hachi: A Dog''s Tale']
movies = []
```

```python
for m in movie_list:
    movie = Movie.objects.search(m)[0]
    #вставляем фильмы в таблицу films
    movie.get_content('main_page')
    movies.append({'title': movie.title,
                   'id' : movie.id,
                   'country' : movie.countries,
                   'box_office' : 0 if movie.profit_world is None else movie.profit_world,
                   'release_year' : movie.year,
                   'actors' : movie.actors,
                   'composers' : movie.composers,
                   'directors' : movie.directors,
                   'editors' : movie.editing_by
                  })
```

```python
p = {}
personid = 0
```

```python
for movie in movies:
    print(f"insert into films VALUES ('{movie['title']}',{movie['id']},'{','.join(movie['country'])}',{movie['box_office']},'{movie['release_year']}-01-01');")
    for x in movie['actors']:
        if (x.name in p) and (not ('actors' in p[x.name][1])):
            p[x.name][1].append('actors')
        else:
            p[x.name] = [personid, ['actors']]
            personid += 1
            print(f"insert into persons VALUES ({personid},'{x.name}');")
            print(f"insert into persons2content VALUES ({personid},{movie['id']},'актер');")
    for x in movie['composers']:
        if (x.name in p) and (not ('composers' in p[x.name][1])):
            p[x.name][1].append('composers')
        else:
            p[x.name] = [personid, ['composers']]
            personid += 1
            print(f"insert into persons VALUES ({personid},'{x.name}');")
            print(f"insert into persons2content VALUES ({personid},{movie['id']},'композитор');")
    for x in movie['directors']:
        if (x.name in p) and (not ('directors' in p[x.name][1])):
            p[x.name][1].append('directors')
        else:
            p[x.name] = [personid, ['directors']]
            personid += 1
            print(f"insert into persons VALUES ({personid},'{x.name}');")
            print(f"insert into persons2content VALUES ({personid},{movie['id']},'режиссер');")
    for x in movie['editors']:
        if (x.name in p) and (not ('editors' in p[x.name][1])):
            p[x.name][1].append('editors')
        else:
            p[x.name] = [personid, ['editors']]
            personid += 1
            print(f"insert into persons VALUES ({personid},'{x.name}');")
            print(f"insert into persons2content VALUES ({personid},{movie['id']},'редактор');")
```
